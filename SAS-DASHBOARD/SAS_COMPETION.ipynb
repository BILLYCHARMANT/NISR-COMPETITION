```python
from dash import Dash, html, dcc
import plotly.express as px
import pandas as pd
from dash.dependencies import Input, Output


app = Dash(__name__)

excel_file = 'summary.xlsx'
sheet='season A'
# sheet1='percentagePerDistrict'
sheet2='Areas'
sheet3='cultivatedArea (season A)'
sheet4='seasons'
df = pd.read_excel(excel_file,sheet_name=sheet)
dfCopy = pd.read_excel(excel_file,sheet_name='copy')

df1 = pd.read_excel(excel_file,sheet_name=sheet2)
df2 = pd.read_excel(excel_file,sheet_name='copy')
df3 = pd.read_excel(excel_file,sheet_name=sheet3)
df4 = pd.read_excel(excel_file,sheet_name=sheet4)


fig = px.bar(df, x="Districts" , y="averageYield", title="Amount in Districts")
fig1 = px.pie(dfCopy, names="Districts" , values="Percentage of farmers who applied organic fertilizer" ,title="Amount in Districts")
fig2 = px.pie(df4, names="season", values="percent",hole=0.4, title="summary of seasons")
fig3 = px.line(df3, x="crops", y="Nyarugenge", title="Amount in Districts")


district_options = [{'label': district, 'value': district} for district in dfCopy['Districts'].unique()]
# column_options = [{'label': column, 'value': column} for column in df.columns]
column_options_one = [{'label': column, 'value': column} for column in df1.columns]
column_options_two = [{'label': column, 'value': column} for column in df2.columns]
column_options_three = [{'label': column, 'value': column} for column in df3.columns]
column_options_four = [{'label': column, 'value': column} for column in df.columns]





# parent div
app.layout = html.Div(children=[
    html.H1(children='AGRICULTURE',style={'color':'dark-blue','text-align':'center'}),
       
    
    html.Div([              
        # first grid
        html.Div([
        #first div
        html.Div([
       html.Div([
           dcc.Dropdown(
        id='dropdown_zero',
        options=column_options_one,
        value=df1.columns[1],
        multi=False,
        style={'width': '40vh','margin-top':'5px'}
    ),
        dcc.Dropdown(
        id='seasonal-dropdown_zero',
        options=[
            {'label': 'Season A', 'value': 'season A'},
            {'label': 'Season B', 'value': 'season B'},
            {'label': 'Season C', 'value': 'season C'},
        ],
        value='season A',
        multi=False,
        style={'width': '40vh','margin-top':'5px'}
    )
       ],style={'display':'flex'}),
           html.Div([
        dcc.Graph(
        id='example-graph-zero',
        figure=fig
        )
           ], style={'width':'49vw','height':'80vh'})], 
        style={'flex-grow':'2','display':'grid'}),
        
#         second div
        html.Div([
         html.Div([
        dcc.Dropdown(
        id='dropdown_one',
        options=column_options_two,
        value=df2.columns[1],
        multi=False,
        style={'width': '40vh','margin-top':'5px'}
    ),
        dcc.Dropdown(
        id='seasonal-dropdown_one',
        options=[
            {'label': 'Season A', 'value': 'season A'},
            {'label': 'Season B', 'value': 'season B'},
            {'label': 'Season C', 'value': 'season C'},
        ],
        value='season A',
        multi=False,
        style={'width': '40vh','margin-top':'5px'}
    )], style={'display':'flex'}),
        html.Div([
        dcc.Graph(
        id='example-graph-one',
        figure=fig1
      )
        ], style={'width':'49vw','height':'80vh'})],
        style={'flex-grow':'2','display':'grid'}),
        ], style={'display':'flex','justify-content':'space-around'}),
#             second grid
        
        html.Div([
#             third div
        html.Div([
       html.Div([
           dcc.Dropdown(
        id='dropdown_default',
        options=column_options_four,
        value=df3.columns[1],
        multi=False,
        style={'width': '40vh','margin-top':'5px'}
    ),
        dcc.Dropdown(
        id='seasonal-dropdown_zer',
        options=[
            {'label': 'Season A', 'value': 'season A'},
            {'label': 'Season B', 'value': 'season B'},
            {'label': 'Season C', 'value': 'season C'},
        ],
        value='season A',
        multi=False,
        style={'width': '40vh','margin-top':'5px'}
    )
       ],style={'display':'flex'}),
           html.Div([
        dcc.Graph(
        id='example-graph-two',
        figure=fig2
        )
           ], style={'width':'49vw','height':'80vh'})], 
        style={'flex-grow':'2','display':'grid'}),
#             fourth div
        html.Div([
        html.Div([ 
            dcc.Dropdown(
        id='dropdown_two',
        options=column_options_three,
        value=df3.columns[1],
        multi=False,
        style={'width': '40vh','margin-top':'5px'}
    ),
        dcc.Dropdown(
        id='seasonal-dropdown_two',
        options=[
            {'label': 'season A', 'value': 'season A'},
            {'label': 'season B', 'value': 'season B'},
            {'label': 'season C', 'value': 'season C'},
        ],
        value='season A',
        multi=False,
        style={'width': '40vh','margin-top':'5px'}
     )], style={'display':'flex'}),
        html.Div([
        dcc.Graph(
        id='example-graph_three',
        style={'color':'red'},
        figure=fig3
        )
        ], style={'width':'49vw','height':'80vh'})],
        style={'flex-grow':'2','display':'grid'}),
        ], style={'display':'flex','justify-content':'space-around'})], 
    ),

])

# first dropdown assigned to a graph
@app.callback(
    Output('example-graph-zero', 'figure'),
    [Input('dropdown_zero', 'value'),
     Input('seasonal-dropdown_zero', 'value')]
)
def update_bar_one(selected_y_axis, selected_season):
    data_frame = pd.read_excel(excel_file, sheet_name=selected_season)
    bar_fig = px.bar(data_frame, x="Districts", y=selected_y_axis,
                     labels={'value': f'{selected_y_axis}'})
    
    bar_fig.update_layout(
        title=f"{selected_y_axis} per District ({selected_season})",
        font=dict(size=9, family='Arial', color='black'),
        title_x=0.5,  # Center the title
        title_font=dict(size=18)
    )
    return bar_fig


# second dropdown assigned to a piechat
@app.callback(
    Output('example-graph-one', 'figure'),
    [Input('dropdown_one', 'value'),
     Input('seasonal-dropdown_one', 'value')]
)
def update_bar_one(selected_y_axis,selected_season):
    frame2 = pd.read_excel(excel_file, sheet_name=selected_season)
    pie_fig = px.pie(frame2, names="Districts", values=selected_y_axis,
#                     title=f"{selected_y_axis} per District",
                    labels={'value': f'{selected_y_axis}'})
    
    pie_fig.update_layout(
    title=f"{selected_y_axis} per District ({selected_season})",
    font=dict(size=9, family='Arial', color='black'),
    title_x=0.5,  # Center the title
    title_font=dict(size=18)
)
    return pie_fig
# third dropdown assigned to graph
@app.callback(
    Output('example-graph_three', 'figure'),
    [Input('dropdown_two', 'value'),
     Input('seasonal-dropdown_two', 'value')]
)
def update_bar_one(selected_y_axis,selected_season):
    new= f"cultivatedArea ({selected_season})"
    frame3 = pd.read_excel(excel_file, sheet_name=new)
    line_fig = px.line(frame3, x="crops", y=selected_y_axis,
                    labels={'value': f'{selected_y_axis}'})
    
    line_fig.update_layout(
    title=f"{selected_y_axis} per District ({selected_season})",
    font=dict(size=9, family='Arial', color='black'),
    title_x=0.5,  # Center the title
    title_font=dict(size=18)
)
    return line_fig



if __name__ == '__main__':
    app.run()
```
